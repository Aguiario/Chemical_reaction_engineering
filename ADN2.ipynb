```python
import numpy as np
import matplotlib.pyplot as plt

# Definir rango de tiempo
t = np.arange(0, 70.1, 0.1)

# Concentración inicial de ADN
S0 = 50

# Vía A
kcat_A = 0.00139
KM_A = 1863
E_A = 48.8
yA = S0 * (1 - np.exp(- (kcat_A * E_A) * t))

# Vía B
kcat_B = 0.008
KM_B = 500
E_B = 19.8
yB = S0 * (1 - np.exp(- (kcat_B * E_B) * t))

# Vía C
kcat_C = 0.0022
KM_C = 67
E_C = 19.8
yC = S0 * (1 - np.exp(- (kcat_C * E_C) * t))

# Crear la gráfica
plt.figure(figsize=(8, 6))
plt.plot(t, yA, 'g-', linewidth=2, label='Vía A')
plt.plot(t, yB, 'r-', linewidth=2, label='Vía B')
plt.plot(t, yC, 'b-', linewidth=2, label='Vía C')

# Configuración de la gráfica
plt.xlabel('Tiempo (s)')
plt.ylabel('Concentración de ADN reparado (nM)')
plt.title('Evolución de la reparación de ADN en distintas vías')
plt.legend()
plt.grid(True)
plt.show()
```
